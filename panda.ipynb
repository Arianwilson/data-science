```python
import pandas as pd
df = pd.DataFrame({
    'student name ': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [20, 22, 19, 21, 23],
    'exam': ['passed', 'failed', 'passed', 'passed', 'failed'],
    'internet access': ['yes', 'no', 'yes', 'yes', 'no'],    
    'gender': ['female', 'male', 'male', 'male', 'female'],
    'presence': ['present', 'absent', 'present', 'present', 'absent'],
    'score' : [18, 8, 15, 13, 7],
    'study hours': [2, 1, 3, 4, 1]
    

})  
```
